04_getTSS.sh　より抽出したTSSのデータを使用して、スクリーニングした遺伝子(全遺伝子も)からTSSの上流2kbを抽出してbed形式のファイルを出力するコード<br>
なお、getTSS.shではgtfからTSSの上流・下流10bを抽出している。<br>
出力は genes/TSS_BDF1_expressed.bed と genes/allgenes_TSS.bed

```python
#! usr/bin/env python
```

```python
#ライブラリの読み込み
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
i = 'BDF1_expressed'
```

```python
i = str(i)
```

```python
#データの読み込み
df = pd.read_csv('./ensembl.gtf.tss.bed', sep='\t',header=None)
df = df.rename(columns = {6:'gene_id'})
df
```

```python
#データの読み込み(trimmed_genes)
ex = pd.read_csv("./genes/"+i+".tsv",sep='\t')
ex
```

```python
bd = pd.merge(df,ex ,on = 'gene_id',how = 'right')
bd = bd.dropna()
bd
```

```python
#上流1990bをとってくる、下流10bを削る
bd.loc[bd[3] == '+', 1] = bd[1] - 1990
bd.loc[bd[3] == '+', 2] = bd[2] - 10

bd.loc[bd[3] == '-', 2] = bd[2] + 1990
bd.loc[bd[3] == '-', 1] = bd[1] + 10
```

```python
bd
```

```python
#データの整形
bd[0] = 'chr' + bd[0]
bd['score'] = 0
bd[1] = bd[1].astype(str)
bd[1] = bd[1].str.replace('\.0','')
bd[2] = bd[2].astype(str)
bd[2] = bd[2].str.replace('\.0','')

bd
```

```python
bd1 = bd[[0,1,2,'gene_id','score',3]]
```

```python
bd1 = bd1.rename(columns = {0:'chr',1:'chromStart',2:'chromEnd',3:'strand'})
bd1
```

```python
bd1.to_csv('./genes/TSS_'+i+'.bed', sep='\t',index=None ,header=None)
```

全遺伝子も

```python
#上流1990bをとってくる、下流10bを削る
df.loc[df[3] == '+', 1] = df[1] - 1990
df.loc[df[3] == '+', 2] = df[2] - 10

df.loc[df[3] == '-', 2] = df[2] + 1990
df.loc[df[3] == '-', 1] = df[1] + 10
```

```python
df[0] = 'chr' + df[0]
df
```

```python
df.to_csv('./genes/allgenes_TSS.bed', sep='\t',index=None ,header=None)
```
