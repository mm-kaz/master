BDF1>ICR>ddY(FPKM値)となる母性因子の候補遺伝子を抽出するためのコード<br>
(候補遺伝子は genes/BDF1_expressed.tsv)

```python
#!/usr/bin/env python
```

*x,a* : BDF1 vs ICR (log2FCが<span style="color: red; ">x(0.8)</span>以上、P-valueが0.05未満、かつBDF1のFPKM値が<span style="color: red; ">a(1)</span>以上となる遺伝子の抽出)　　<br>
*y,b* : ICR vs ddY (log2FCが<span style="color: red; ">y(0.1)</span>以上、P-valueが0.05未満、かつICRのFPKM値が<span style="color: red; ">b(0.1)</span>以上となるとなる遺伝子の抽出)

```python
i = 'BDF1_expressed'

x = 0.8
y = 0.1

a = 1
b = 0.1
```

```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#データの読み込み DEseq2
Bvd = pd.read_csv('./DEseq2_data/BDF1vsddY_lrt.tsv',sep='\t')
BvI = pd.read_csv('./DEseq2_data/BDF1vsICR_lrt.tsv',sep='\t')
Ivd = pd.read_csv('./DEseq2_data/ICR1vsddY_lrt.tsv',sep='\t')
```

```python
#gene_idを列に変換
Bvd['gene_id'] = Bvd.index
BvI['gene_id'] = BvI.index
Ivd['gene_id'] = Ivd.index
```

```python
#発現量データの読み込み
ex = pd.read_csv('./seqData/M2_expression_mm10.tsv',sep='\t')
```

```python
#データの結合 & NAを含む行の削除
Bvd = pd.merge(Bvd,ex ,on = 'gene_id',how = 'outer')
Bvd = Bvd.dropna(how='any')

BvI = pd.merge(BvI,ex ,on = 'gene_id',how = 'outer')
BvI = BvI.dropna(how='any')

Ivd = pd.merge(Ivd,ex ,on = 'gene_id',how = 'outer')
Ivd = Ivd.dropna(how='any')
```

```python
Bvd
```

```python
num1 = BvI[BvI['BDF1_FPKM_avr'] >= a]
num2 = Ivd[Ivd['ICR_FPKM_avr'] >= b]
num = pd.merge(num1,num2 ,on = 'gene_id',how = 'inner')
num
```

```python
#BDF1 vs ICR (log2FCがx以上、P-valueが0.05未満、かつBDF1のFPKM値が0.5以上となる遺伝子の抽出)
B_I = BvI[(BvI['log2FoldChange'] >= x) & (BvI['pvalue'] < 0.05) & (BvI['BDF1_FPKM_avr'] >= a)]
```

```python
#ICR vs ddY (log2FCがy以上、P-valueが0.05未満となる遺伝子の抽出)
I_d = Ivd[(Ivd['log2FoldChange'] >= y) & (Ivd['pvalue'] < 0.05) & (Ivd['ICR_FPKM_avr'] >= b)]
```

```python
#BDF1
B_sp = pd.merge(B_I,I_d ,on = 'gene_id',how = 'inner')
B_sp = B_sp.rename(columns={'BDF1_FPKM_avr_y': 'BDF1_FPKM_avr','ICR_FPKM_avr_y': 'ICR_FPKM_avr','ddY_FPKM_avr_y': 'ddY_FPKM_avr'})
```

```python
B_sp = B_sp.set_index('gene_id')
B_sp = B_sp[['BDF1_FPKM_avr','ICR_FPKM_avr','ddY_FPKM_avr']]
```
