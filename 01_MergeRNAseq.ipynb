```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#データの読み込み rep1
BDF1_1 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0028_D04_Dr_Funaya_1_BDF1_MII_S28_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0028_D04_Dr_Funaya_1_BDF1_MII_S28_L002.genes.results',sep='\t')
ICR_1 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0029_E04_Dr_Funaya_1_ICR_MII_S29_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0029_E04_Dr_Funaya_1_ICR_MII_S29_L002.genes.results',sep='\t')
ddY_1 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0030_F04_Dr_Funaya_1_ddY_MII_S30_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0030_F04_Dr_Funaya_1_ddY_MII_S30_L002.genes.results',sep='\t')
```

```python
#データの読み込み rep2
BDF1_2 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0031_G04_Dr_Funaya_2_BDF1_MII_S31_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0031_G04_Dr_Funaya_2_BDF1_MII_S31_L002.genes.results',sep='\t')
ICR_2 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0032_H04_Dr_Funaya_2_ICR_MII_S32_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0032_H04_Dr_Funaya_2_ICR_MII_S32_L002.genes.results',sep='\t')
ddY_2 = pd.read_csv('/Volumes/HDD1/m2RNA-seq/210426_NovaSeq_SP_TruseqUD_l2_0033_A05_Dr_Funaya_2_ddY_MII_S33_L002/mm10/rsem/210426_NovaSeq_SP_TruseqUD_l2_0033_A05_Dr_Funaya_2_ddY_MII_S33_L002.genes.results',sep='\t')
```

```python
#DataFrame作成
df = pd.DataFrame(index=[], columns=[])
df['gene_id'] = BDF1_1['gene_id']
```

```python
#TPM
df['BDF1_TPM_avr'] = (BDF1_1['TPM']+BDF1_2['TPM']) / 2
df['ICR_TPM_avr'] = (ICR_1['TPM']+ICR_2['TPM']) / 2
df['ddY_TPM_avr'] = (ddY_1['TPM']+ddY_2['TPM']) / 2
```

```python
#FPKM
df['BDF1_FPKM_avr'] = (BDF1_1['FPKM']+BDF1_2['FPKM']) / 2
df['ICR_FPKM_avr'] = (ICR_1['FPKM']+ICR_2['FPKM']) / 2
df['ddY_FPKM_avr'] = (ddY_1['FPKM']+ddY_2['FPKM']) / 2
```

```python
df
```

```python
df.to_csv("./seqData/M2_expression_mm10.tsv",index=False,sep='\t')
```
