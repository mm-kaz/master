Metascapeを用いて候補遺伝子を対象にGO解析を行った。<br>
条件は<br>
Min Overlap:3<br>
P Value Cutoff:0.05<br>
Min Enrichment:1.5<br>
データベースはMolecular Functions, Cellular Components, Biological Processes を使用した。<br>
出力データの全てをダウンロードした。<br>
以上の結果を可視化するためのコード

```python
#! usr/bin/env python
```

```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline

# 目盛り調整のticker追加
import matplotlib.ticker as ticker
```

```python
#データの読み込み
df = pd.read_csv('./metascape/Enrichment_GO/_FINAL_GO.csv',sep=',')
df
```

**Molecular Functions**

```python
MF = df[df['Category'] == 'GO Molecular Functions']
MF = MF.sort_values(by="_LogP_MyList")
MF["_LogP_MyList"] = MF["_LogP_MyList"] * -1
MF
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 5))
ax.set_xlim(0,3.2)

plt.tick_params(labelsize=13)

# Horizontal Bar Plot
ax.barh(MF['Description'], MF['_LogP_MyList'],color ='lightcoral')

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.4,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
    
    
#Add Plot Title
ax.set_title('GO Molecular Functions',loc ='center', fontsize =17)

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

# Add Text watermark
fig.text(0.5, 0.02, '- log10(p-value)', fontsize =14,
         color ='black', ha ='center', va ='bottom', alpha = 0.7)

# Add Text watermark
fig.text(0.8, 0.87, 'n=94', fontsize =14,
         color ='black', ha ='center', va ='bottom', alpha = 0.7)
 
# save Plot
plt.savefig("./outfig/SCgenes_GO-MF.png", dpi=600, bbox_inches="tight", pad_inches=0.1)
```
